### Ordem em Listas
- Método .sort()
- Argumento opcional: Reverse

```python
import pandas as pd

dados = ["Ana", "Paulo", "Felipe"]

# dados.sort() # Ordem alfabética
dados.sort(reverse=True) # Ordem alfabética reversa

dados
```
