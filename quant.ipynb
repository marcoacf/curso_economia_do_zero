### Python para análise quantitativa
- Biblioteca Quantstats
- Pegar os dados (yfinance)
- Métricas estatísticas (stats)
- Gráficos complexos (plots)
- Exportar reports

```python
import quantstats as qs, yfinance as yf

ativo = yf.download("BITH11.SA")["Adj Close"]

ativo

benchmark = yf.download("BOVA11.SA")["Adj Close"]
benchmark
```

```python
qs.stats.cagr(ativo) # CAGR = crescimento anual composto
qs.stats.max_drawdown(ativo) # Drawdown = quedas
qs.stats.volatility(ativo) # Volatilidade = desvio-padrão
qs.stats.value_at_risk(ativo) #Value at Risk = gestão de risco máximo esperado
qs.stats.monthly_returns(ativo) # EOY = End of Year
qs.stats.sharpe(ativo) # Sharpe = métrica importantíssima para distinguir fundos
```
