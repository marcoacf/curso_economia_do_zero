### Localizando dados
- Carregar os dados (argumento "index_col")
- Excluir as colunas indesejadas (argumento "inplace=True")
- Usar métodos .loc e .iloc (utilizam [] e não parênteses)
- Segmentação cirúrgica dos dados
- Axis (Eixos): 0 (linha) ou 1 (coluna)

```python
import pandas as pd

dados = pd.read_excel("CVM.xlsx", index_col="DATA")
dados.drop("Unnamed: 0", axis=1, inplace=True) # i, j (Python: 0 ou 1)

dados
```

Método .loc[]

```python
dados.loc["2024-01-02", ["COTISTAS", "PL"]]
```

Método .iloc[]

```python
dados.iloc[0:5, 4:6]
```
