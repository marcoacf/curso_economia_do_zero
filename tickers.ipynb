```python
import yfinance as yf
from concurrent.futures import ThreadPoolExecutor

def get_stats(ticker):
    info = yf.Tickers(ticker).tickers[ticker].info
    print(f"{ticker} - Preço: {info['currentPrice']} - Valor de Mercado:{round(info['marketCap']/1000000,1)} - margem bruta:{round(info['grossMargins']*100,1)}% - margem EBTIDA: {round(info['ebitdaMargins']*100,1)}% - dividendos:{info['dividendYield']} ")

```

```python
ticker_list = ['AAPL', 'ORCL', 'MSFT', 'AMZN', 'GOOG']

with ThreadPoolExecutor() as executor:
    executor.map(get_stats, ticker_list)
```
