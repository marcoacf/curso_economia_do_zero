### Dados macroeconômicos
- Biblioteca "python-bcb"
- Extração de dados
- Tratamento de dados
- Gráficos ilustrativos

**Exemplos:**   
24369 - Desemprego   
432 - Selic   
21774 - População   
27810 - M2 (Agregado monetário)

```python
from bcb import sgs
import pandas as pd
```

```python
desemprego = sgs.get(("Desemprego", 24369))

selic = sgs.get(("Selic", 432))

população = sgs.get(("População", 21774))

m2 = sgs.get(("M2", 27810))
m2["M2"] = pd.to_numeric(m2["M2"])
m2.dropna(inplace=True)
```

```python
desemprego.plot(
    title="Desemprego",
    color="green",
    xlabel="Datas"
)
```
