### Câmbio e criptomoedas
- Biblioteca yfinance
- Dólares (USD) e Euros (EUR)

```python
import yfinance as yf, pandas as pd

dados = yf.download("EURBRL=X")["Close"]

dados.plot(title="Euros")
```

```python
bitcoin = yf.download("BTC-USD")["Close"]

bitcoin.plot(title="Bitcoin", color="yellow")
```

```python
pd.concat([dados, bitcoin], axis=1).dropna().plot()
```
